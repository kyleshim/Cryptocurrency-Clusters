```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Read `crypto_data.csv` into Pandas
file_path = Path("crypto_data.csv")
df = pd.read_csv(file_path, index_col=0)
df
```

```python
# Discard all cryptocurrencies that are not being traded
df['IsTrading'].value_counts()
```

```python
trading_df = df[df['IsTrading'] == True]
trading_df
```

```python
# Remove all rows that have at least one null value
trading_df.isnull().sum()
```

```python
trading_df = trading_df.dropna()
trading_df
```

```python
trading_df = trading_df.drop(columns=['IsTrading'], axis=1)
trading_df
```

```python
# Filter for cryptocurrencies that have been mined
mined_df = trading_df[trading_df['TotalCoinsMined'] > 0]
mined_df
```

```python
# Delete the `CoinName` from the original dataframe
nameless_df = mined_df.drop(columns=['CoinName'], axis=1)
nameless_df
```

```python
# Convert the remaining features with text values, `Algorithm` and `ProofType`, into numerical data
num_df = pd.get_dummies(nameless_df, columns=['Algorithm','ProofType'])
num_df
```

The dataframe expanded from 4 columns to 98 columns with the numerical conversion of Algorithm and Proof Type 

```python
# Standardize your dataset
scaler = StandardScaler()
scaled_data = scaler.fit_transform(num_df)
```

```python
# Initialize PCA model
pca = PCA(n_components=0.9)

# Get two principal components for the iris data.
scaled_pca = pca.fit_transform(scaled_data)
```

```python
pca_df = pd.DataFrame(data=scaled_pca, index=num_df.index)
pca_df
```

The number of features was reduced from 98 to 74 columns in order preserve 90% explained variance

```python
# Initialize t-SNE model
tsne = TSNE(learning_rate=200)
# Reduce dimensions
tsne_features = tsne.fit_transform(pca_df)
# The dataset has 2 columns
tsne_features.shape
```

```python
tsne_df = pd.DataFrame(data=tsne_features, columns=['Principal Component 1','Principal Component 2'])
tsne_df
```

```python
# Visualize the clusters
plt.scatter(tsne_df['Principal Component 1'], tsne_df['Principal Component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

There are at least 2 and maybe 3 distinct cluster after the tsne transformation

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(tsne_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for customer data')
plt.show()
```

```python
# Initialize the K-Means model
model = KMeans(n_clusters=4, random_state=42)

# Train the model
model.fit(tsne_df)

# Predict clusters
predictions = model.predict(tsne_df)

# Create return DataFrame with predicted clusters
tsne_df["class"] = model.labels_
```

```python
plt.scatter(tsne_df['Principal Component 1'], tsne_df['Principal Component 2'], c=tsne_df['class'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# Recommendations

Based on the findings it does appear that cryptocurrency can be clustered. The findings show 4 clusters. However, there is one outlier and a large amount of data lost in the processing step. This leads to believe that the best course forward would be to collect more data and re-analyze.
